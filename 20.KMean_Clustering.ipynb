```python
# KMean clustering
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

df = pd.read_csv("DataSets/ML19_DATA.csv")
df
```

```python
plt.scatter(df.Age ,df.Income)
```

```python
km = KMeans(n_clusters = 3)
```

```python
km.fit_predict(df[['Age','Income']])
```

```python
df['cluster'] = km.fit_predict(df[['Age','Income']])
df
```

```python
df0 = df[df.cluster == 0]
df1 = df[df.cluster == 1]
df2 = df[df.cluster == 2]
```

```python
plt.scatter(df0.Age , df0.Income)
plt.scatter(df1.Age , df1.Income)
plt.scatter(df2.Age , df2.Income)
# This is not proper clustering so we scale tha data
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
mms.fit(df[['Age']])
df['Age'] = mms.transform(df[['Age']])
mms.fit(df[['Income']])
df['Income'] = mms.transform(df[['Income']])
# Minmax scaler convert scale values in range 0 to 1
```

```python
df
```

```python
km.fit_predict(df[['Age','Income']])
```

```python
df['cluster'] = km.fit_predict(df[['Age','Income']])
df
```

```python
df0 = df[df.cluster == 0]
df1 = df[df.cluster == 1]
df2 = df[df.cluster == 2]
```

```python
plt.scatter(df0.Age , df0.Income)
plt.scatter(df1.Age , df1.Income)
plt.scatter(df2.Age , df2.Income)
 
```

```python
centroid = km.cluster_centers_
centroid
```

```python
c0 = centroid[: ,0]  # 1st col all rows
c1 = centroid[: ,1]  # 2st col all rows
 
```

```python
plt.scatter(df0.Age , df0.Income)
plt.scatter(df1.Age , df1.Income)
plt.scatter(df2.Age , df2.Income)
plt.scatter(c0,c1, color = 'black',marker = '*')
```

```python
k_range = range(1,9)
sse = []
for k in k_range:
    km = KMeans(n_clusters=k)
    km.fit(df[['Age','Income']])
    sse.append(km.inertia_)       # Here inertia_ giver sum of square error (square of dist. bet. centroids and datapoints)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('SSE')
plt.plot(k_range,sse)
```
