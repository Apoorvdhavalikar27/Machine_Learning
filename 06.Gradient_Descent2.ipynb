```python
import pandas as pd
import math
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = (10,5)

df = pd.read_csv("DataSets/ML5_DATA.csv")
df
```

```python
x = df.math
y = df.cs
```

```python
plt.scatter(x,y,marker='*',color = 'red',s = 100)
```

```python
# Here we use gradient Descent algo to find the relation in math(x) and cs(y)
 
def gradient_descent(x,y):
    
    m_curr = b_curr = 0
    iteration = 1000 
    n = len(x)
    learning_rate = 0.0002
    
    for i in range(iteration):
        y_predicted = m_curr * x + b_curr
        cost = (1/n)*sum([val**2 for val in (y-y_predicted)])
        dm = -(2/n)*sum(x*(y-y_predicted))
        db = -(2/n)*sum(y-y_predicted)
        m_curr = m_curr - learning_rate * dm
        b_curr = b_curr - learning_rate * db
        plt.plot(x,y_predicted,color = 'green')
        plt.scatter(x,y,marker='*',color = 'red',s = 100)
        print("m {},b {},cost {} ,iteration {}".format(m_curr,b_curr,cost,i)) 
```

```python
gradient_descent(x,y)
```

```python
def gd(x,y):
    
    m = b= 0
    lr = 0.0002
    n = len(x)
    epo = 1000
    
    for i in range(epo):
        y_predicted = m*x + b
        cost = (1/n)*sum([s**2 for s in (y-y_predicted)])
        dm = -(2/n)*sum(x*(y-y_predicted))
        db = -(2/n)*sum(y-y_predicted)
        m = m - lr*dm
        b = b - lr*db
        plt.plot(x,y_predicted,color = 'green')
        plt.scatter(x,y,marker='*',color = 'red',s = 100)
        print("m {},b {},cost {} ,iteration {}".format(m,b,cost,i)) 
```

```python
gd(x,y)
```
