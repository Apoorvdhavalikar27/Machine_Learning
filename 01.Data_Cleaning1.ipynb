```python
def set_area(x):
    try:
            token1,token2 = float(x.split('-')[0]),float(x.split('-')[1])
            area = (token1+token2)/2
            return area
        
    except:
            try:
                x = float(x)
                return x
            
            except:
                return True
         
```

```python
# Now we remove data which are beyond certain standard deviation use some function
# aps - area per sqft
def remove_aps_outliers(df):
    new_df = pd.DataFrame()
    for key,buffer_df in df.groupby('location'):
        m = np.mean(buffer_df.area_per_sqft)
        sd = np.std(buffer_df.area_per_sqft)
        reduced_df = buffer_df[(buffer_df.area_per_sqft>(m-sd)) & (buffer_df.area_per_sqft<=(m+sd))]
        new_df = pd.concat([new_df,reduced_df] ,ignore_index = True)
    return new_df
```

```python
df6 = remove_aps_outliers(df5)
df6.shape
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = (15,10)
def plot_scatter(df,location):
    bhk2 = df[(df.BHK == 2) & (df.location ==location)]
    bhk3 = df[(df.BHK == 3) & (df.location ==location)]
#   matplotlib.rcParams['figure.figsize'] = (15,10)
    plt.scatter(bhk2.area,bhk2.area_per_sqft,color='blue',label='2 BHK', s = 50)
    plt.scatter(bhk3.area,bhk3.area_per_sqft,marker='+',color='green',label='3 BHK', s = 50)
    plt.xlabel('Area')
    plt.ylabel('Area_per_sqft')
    plt.title('Location')
    plt.legend()
```

```python
plot_scatter(df6,'Rajaji Nagar')
# Here some 2bhk points value having higher price than 3bhk points
```
